9.) The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into
suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where
size='Varies with device')

```python
ara.Size.unique()
```

```python
def cal_value(x):
    if 'M' in x:
        x = x.replace('M','')
        return float(x)*1000000
    elif 'k' in x:
        x = x.replace('k','')
        return float(x)*1000
    elif x.isnumeric():
        return float(x)
    else:
        return np.nan
```

```python
ara['Size'] = ara['Size'].apply(cal_value)
ara['Size']
```

```python
ara.dropna(subset = ['Size'],inplace = True)
ara['Size'].isnull().sum()
```

```python
ara['Size'].unique()
```
